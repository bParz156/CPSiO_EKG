```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np

wp=0.2
ws=0.3
gpass=3
gstop=40

frequency=360
halfFreq=frequency/2
stopband=60
order=10
fmin=0
fmax=frequency
nazwa='noise'
n=3201
T=1/frequency
t = np.linspace(0, (n-1)*T , num=n)
x1=np.sin(2 * np.pi * 50 * t)
x2=np.sin(2 * np.pi * 70 * t)
x=x1+x2

#fmin=0
#fmax=stopband
#f=2*np.pi*np.linspace(fmin, fmax, n)

margin = 0.05  # 5% margin
xmin, xmax = 0, halfFreq  # Desired domain
xmargin = (xmax - xmin) * margin
margins=(xmin - xmargin, xmax + xmargin)

N, Wn = signal.buttord(wp, ws, gpass, gstop, analog=True)
b, a = signal.butter(N, Wn, 'low', analog=True)
w, h = signal.freqs(b, a, n) #, f)

plt.plot(w, (abs(h)))
plt.title('Odpowiedz filtra dolnoprzepustowego Butterwortha')
plt.xlabel('Częstotliwość [Hz]')
plt.ylabel('Amplituda [dB]')
plt.grid(which='both', axis='both')
plt.show()

plt.plot(w, 20*np.log10(abs(h)))
plt.title('Odpowiedz filtra dolnoprzepustowego Butterwortha (skala logarytmiczna)')
plt.xlabel('Częstotliwość [Hz]')
plt.ylabel('Amplituda [dB]')
plt.grid(which='both', axis='both')
plt.show()

```

```python
"""
x to sygnal oryginalny
h to filtr
nazwaFiltra sluzy do opisu wykresu
"""
def filtrDolnoprzepustowy(x,h, nazwaFiltra, frequency=frequency, margins=(-frequency/2, frequency/2)):
    fourier=np.fft.fft(x)
    freq_axis = np.fft.fftfreq(n, d=1/frequency)
    modul=abs(fourier)
    filtrowany=modul*h
    roznicaF=fourier=filtrowany
    
    plt.figure()
    plt.plot(freq_axis, modul,label='sygnal oryginalny')
    plt.title('Widmo amplitudowe sygnału')
    plt.plot(freq_axis,filtrowany, label='po filtracji: '+nazwaFiltra)
    plt.legend(loc='upper right')
    plt.xlabel('Częstotliwość [Hz]')
    plt.ylabel('Amplituda')
    plt.xlim(margins)

    plt.figure()
    plt.plot(roznicaF)
    plt.title('Roznica w widmie amplitudowym')
    plt.xlabel('Częstotliwość [Hz]')
    plt.ylabel('Amplituda')
    plt.xlim(margins)
    plt.ylim(-10,10)
    
    wyfiltorwany=np.fft.ifft(filtrowany)
    plt.figure()
    plt.plot(wyfiltorwany, color='g', label='sygnal po filtracji '+nazwaFiltra)
    plt.plot(x, color='b', label='sygnal oryginalny')
    plt.title('Po filtracji '+ nazwa)
    plt.xlabel('Czas [s]')
    plt.ylabel('Amplituda')
    plt.legend(loc='upper right')
    plt.xlim(margins)

    roznicaSygnal=x-wyfiltorwany
    plt.figure()
    plt.plot(roznicaSygnal)
    plt.title('Roznica w dziedzinie czasu ')
    plt.xlabel('Czas [s]')
    plt.ylabel('Amplituda')
    plt.xlim(margins)
    
    plt.show()

filtrDolnoprzepustowy(x,h, 'Butteroworth low 60 Hz', frequency=frequency, margins=margins)
```
